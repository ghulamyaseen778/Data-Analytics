```python
import csv
```

```python
open_file = open("./student data.csv",mode="r")
print(open_file)
```

```python
file_reader = csv.reader(open_file)
print(file_reader)
```

```python
data = list(file_reader)
print(data)
```

```python
headers = data[0]
headers
```

```python
student_data = data[1:]
```

```python
chem = []
for i in student_data:
    chem.append(int(i[2]))

print(f"Chemistry Avarage Marks: {sum(chem)//len(chem)}")
```

```python
Physics = []
for i in student_data:
    Physics.append(int(i[3]))

print(f"Physics Avarage Marks: {sum(Physics)//len(Physics)}")
```

```python
Maths = []
for i in student_data:
    Maths.append(int(i[4]))

print(f"Maths Avarage Marks: {sum(Maths)//len(Maths)}")
```

```python
getting_percentage = []
total_marks = 300
for i in student_data:
    data = i
    sum_of_total = (int(i[2]))+int(i[3])+(int(i[4]))
    percentage = (sum_of_total/total_marks*100)
    data.append(round(percentage,2))
    getting_percentage.append(data)
headers.append("Percantage")
getting_percentage.insert(0,headers)
print(getting_percentage)
    
```
